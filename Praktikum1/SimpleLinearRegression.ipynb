```python
import pandas as pd

data= {'luas': [6,8,10,14,18],
       'harga': [7,9,13,17.5,18]}

df=pd.DataFrame(data)
df
```

```python
# prompt: buatkan saya visualisasi data dari data di atas dengan menggunakan scatter plot dengan x sebagai luas dan y sebagai harga, dan judulnya adalah perbandingan luas tanah dengan harga tanah, kemudian xlabel adalah luas dan ylabel adalah harga, dengan grid true

import matplotlib.pyplot as plt

# Create the scatter plot
df.plot(kind='scatter', x='luas', y='harga')
plt.title('Perbandingan Luas Tanah dengan Harga Tanah')
plt.xlabel('Luas tanah (m persegi)')
plt.ylabel('Harga tanah (dollar)')
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

```python
import numpy as np

x=np.array(df['luas'])
y=np.array(df['harga'])

print(f'x: {x}')
print(f'y: {y}')
```

```python
x=x.reshape(-1,1)
print(f'x shape: {x.shape}')
print(f'x: {x}')
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x,y)
```

```python
x_vis=np.array([0,25]).reshape(-1,1)
y_vis=model.predict(x_vis)
```

```python
plt.scatter(x,y)
plt.plot(x_vis,y_vis,c='r')
plt.title('Perbandingan Luas Tanah dengan Harga Tanah')
plt.xlabel('Luas tanah (m persegi)')
plt.ylabel('Harga tanah (dollar)')
plt.grid(True)
plt.show()
```

```python
print(f'intercept: {model.intercept_}')
print(f'slope: {model.coef_}')
```

```python
luas_tanah=np.array([12,15,20]).reshape(-1,1)
print(luas_tanah.shape)
luas_tanah
```

```python
harga_prediksi=model.predict(luas_tanah)
harga_prediksi
```

```python
for luas, hrg in zip(luas_tanah, harga_prediksi):
    print(f'luas tanah: {luas}, harga prediksi: {hrg}')
```

### Evaluasi Simple Linear Regression

Training and testing dataset

```python
x_train = np.array([6,8,10,14,18]).reshape(-1,1)
y_train = np.array([7,9,13,17.5,18])

x_test = np.array([9,9,11,16,12]).reshape(-1,1)
y_test = np.array([11,8.5,15,18,11])
```

Training simple linear regression

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score

y_pred=model.predict(x_test)
r_squared=r2_score(y_test,y_pred)
print(f'R-squared: {r_squared}')
```
